```python
def extract_features(precursor_df, fragment_intensity_df, fragment_mz_df):
    # Store feature vectors in a list
    feature_vectors = []
    identifiers = []  # mod_seq_charge_hash storage
    
    # Convert fragment intensity & m/z data to NumPy arrays (FAST indexing)
    frag_intensity_np = fragment_intensity_df.to_numpy()
    frag_mz_np = fragment_mz_df.to_numpy()
    
    for _, precursor in precursor_df.iterrows():
        mod_seq_charge_hash = precursor["mod_seq_charge_hash"]
        charge = precursor["charge"]
        precursor_mz = precursor["precursor_mz"]
        rt_pred = precursor["rt_pred"]
        mobility_pred = precursor["mobility_pred"]
        frag_start_idx = precursor["frag_start_idx"]
        frag_stop_idx = precursor["frag_stop_idx"]
        
        # Extract fragments from NumPy array (FAST)
        frag_intensities = frag_intensity_np[frag_start_idx:frag_stop_idx, :]
        frag_mz_values = frag_mz_np[frag_start_idx:frag_stop_idx, :]
        
        # Flatten and separate b-ion and y-ion fragments
        b_mz = frag_mz_values[:, :2].ravel()  # 'b_z1', 'b_z2'
        y_mz = frag_mz_values[:, 2:].ravel()  # 'y_z1', 'y_z2'
        b_intensity = frag_intensities[:, :2].ravel()
        y_intensity = frag_intensities[:, 2:].ravel()
        
        # Combine b-ion and y-ion fragments
        all_mz = np.concatenate([b_mz, y_mz])
        all_intensity = np.concatenate([b_intensity, y_intensity])
        ion_types = np.concatenate([np.ones_like(b_mz), np.zeros_like(y_mz)])  # 1 for b-ion, 0 for y-ion
        
        # Find top 25 fragments by intensity
        top_k = min(25, len(all_intensity))
        top_indices = np.argpartition(all_intensity, -top_k)[-top_k:]
        top_frag_mz = all_mz[top_indices]
        top_frag_intensities = all_intensity[top_indices]
        top_ion_types = ion_types[top_indices]
        
        # Sort by fragment_mz (descending order)
        sorted_indices = np.argsort(-top_frag_mz)
        top_frag_mz = top_frag_mz[sorted_indices]
        top_frag_intensities = top_frag_intensities[sorted_indices]
        top_ion_types = top_ion_types[sorted_indices]
        
        
        # Sort by fragment_mz (descending order)
        # sorted_indices = np.argsort(-all_mz)
        # all_mz = all_mz[sorted_indices]
        # all_intensity = all_intensity[sorted_indices]
        # ion_types = ion_types[sorted_indices]
        
        
        # # Step 1: Stack arrays together
        # frag_array = np.stack((all_mz, all_intensity, ion_types), axis=1)

        # # Step 2: Filter by frag_mz range [400, 1000]
        # min_mz = 400
        # max_mz = 1000
        # mask = (frag_array[:, 0].astype(float) >= min_mz) & (frag_array[:, 0].astype(float) <= max_mz)
        # frag_array_filtered = frag_array[mask]
        
        # # Step 1: Convert frag_mz and intensity to float
        # frag_mz = frag_array_filtered[:, 0].astype(float)
        # frag_intensity = frag_array_filtered[:, 1].astype(float)

        # # Step 2: Define bin size
        # bin_size = 5

        # # Step 3: Create bin edges
        # bins = np.arange(min_mz, max_mz + bin_size, bin_size)  # +bin_size to include max_mz

        # # Step 4: Digitize frag_mz into bins
        # bin_indices = np.digitize(frag_mz, bins) - 1  # subtract 1 to make it zero-based indexing

        # # Step 5: Sum intensities inside each bin
        # bin_sums = np.zeros(len(bins))
        # for bin_idx, intensity in zip(bin_indices, frag_intensity):
        #     if 0 <= bin_idx < len(bins):
        #         bin_sums[bin_idx] += intensity
        
        # # print(bins)
        # # print(bin_sums)
        # # print(bin_sums.shape)
        
        
        
        
        # Pad with NaN if fewer than 25 fragments
        pad_size = 25 - len(top_frag_mz)
        if pad_size > 0:
            top_frag_mz = np.pad(top_frag_mz, (0, pad_size), constant_values=np.nan)
            top_frag_intensities = np.pad(top_frag_intensities, (0, pad_size), constant_values=np.nan)
            top_ion_types = np.pad(top_ion_types, (0, pad_size), constant_values=np.nan)
        
        # Create feature vector
        feature_vector = [charge, precursor_mz, rt_pred, mobility_pred] + list(top_frag_mz) + list(top_frag_intensities) + list(top_ion_types)
        
        identifiers.append(mod_seq_charge_hash)
        feature_vectors.append(feature_vector)
    
    # Define column names
    columns = ["charge", "precursor_mz", "rt_pred", "mobility_pred"] + \
              [f"frag_mz_{i}" for i in range(1, 26)] + \
              [f"frag_intensity_{i}" for i in range(1, 26)] + \
              [f"frag_ion_type_{i}" for i in range(1, 26)]
    
    # Convert to DataFrame
    feature_df = pd.DataFrame(feature_vectors, columns=columns)
    feature_df.index = identifiers  # Set mod_seq_charge_hash as index
    
    return feature_df
```

```python
def store_feature_vectors(feature_df,storage_format,output_path):
    # Store feature vectors
    if storage_format == "csv":
        feature_df.to_csv(output_path, index=True, index_label="mod_seq_charge_hash")
    elif storage_format == "hdf5":
        feature_df.to_hdf(output_path, key="features", mode="w")
    else:
        raise ValueError("Invalid storage format. Choose 'csv' or 'hdf5'.")

    print(f"Feature vectors stored in {output_path} as {storage_format.upper()}.")
```
